```python
def train_model(enviroment_name, algorithm, policy, n_envs, n_stack, seed, verbose,total_timesteps):
    env = gym.make(enviroment_name)
    env.reset()
    
    log_path = f"logs/{algorithm}_{policy}_{total_timesteps}_learningrate001"
    model_save_path = f"models/{algorithm}_{policy}_{total_timesteps}_learningrate001"
    
    
    env = make_atari_env(enviroment_name, n_envs=n_envs, seed= seed)
    env = VecFrameStack(env, n_stack=n_stack)
    
    if (algorithm == 'A2C'):
        model = A2C(policy, env, verbose=verbose, learning_rate=0.001,tensorboard_log=log_path)
    if (algorithm == 'PPO'):
        model = PPO(policy, env, verbose=verbose, tensorboard_log=log_path)
    if (algorithm == 'DQN'):
        model = DQN(policy, env, verbose=verbose, tensorboard_log=log_path)
    if (algorithm == 'TD3'):
        model = TD3(policy, env, verbose=verbose, tensorboard_log=log_path)
        

    model.learn(total_timesteps=total_timesteps)
    model.save(model_save_path)
    del model
```
